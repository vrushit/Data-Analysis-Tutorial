```python
#Dealing with missing Data
# Drop the whole row or column
#replace it by mean, frequency 
```

```python
# For column "normalized losses"
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#Replace "NaN" by mean value in "normalized-losses" column
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#Same for Column Bore
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
#For Stroke
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average for stroke:", avg_stroke)
```

```python
df['stroke'].replace(np.nan, avg_stroke, inplace= True)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df['num-of-doors'].value_counts() #See what values are present in what numbers
```

```python
#Most Common 
df['num-of-doors'].value_counts().idxmax()
```

```python
#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()
```

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()

#The above is just for example the columns length, width are a part for our dataset
```

```python
#Normalizing Column Height

df['height'] = df['height']/df['height'].max()
df[["length","width","height"]].head()
```

```python
# ANOVA: Analysis of Variance
# important
```

```python
#F-test score: ANOVA assumes the means of all groups are the same, calculates how much the actual means deviate from the assumption, and reports it as the F-test score. A larger score means there is a larger difference between the means.
#P-value: P-value tells how statistically significant is our calculated score value.
```

```python
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

```python
df_gptest
```

```python
grouped_test2.get_group('4wd')['price']
```

```python
# ANOVA
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

```python
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('rwd')['price'])  
   
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```
